```
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense, Activation
import pandas as pd
from sklearn.model_selection import train_test_split
```

```
df = pd.read_csv('winequality-red.csv')
df.head()
```

```
df.tail()
```

```
df.size
```

```
df.shape
```

```
X = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides','free sulfur dioxide','total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol' ]]
y = df[['quality']]
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
y_train.shape
```

```

# define the keras model
model = Sequential()
model.add(Dense(11, input_dim=11, kernel_initializer='normal',activation='relu'))
model.add(Dense(6, kernel_initializer='normal',activation='relu'))
model.add(Dense(4, kernel_initializer='normal',activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
```

```
# Compilaci√≥n del modelo
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['mean_absolute_percentage_error'])
```

```
#Entrena el modelo
x_val = X_train[840:]
y_val = y_train[840:]
#Entrena el modelo
model.fit(X_train, y_train, batch_size=32, epochs=200,validation_data=(x_val,y_val))
# batch_size cantidad de datos que vamos suministrando para el entrenamiento
# epochs, cantidad de iteracion que hace el aalgoritmo para ir aprendiendo
```

```
results = model.evaluate(X_test, y_test)

for i in range(len(model.metrics_names)):
  print(model.metrics_names[i]," : ", results[i])
```
